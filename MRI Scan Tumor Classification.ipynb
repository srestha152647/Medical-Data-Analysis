```python
x_train2, x_val, y_train2, y_val = train_test_split(x_train, y_train, test_size = 0.1, random_state = 3)
```

### VGG16 Model - Transfer Learning

```python
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from keras.optimizers import Adam
# Transfer Learning Model
from tensorflow.keras.applications import VGG16
```

```python
def build_model():
    #use Imagenet = pre-trained models weights called knowledge that is transferred 
    # image_shape = 32x32x3
    vgg16_model = VGG16(weights= 'imagenet', include_top= False, input_shape=(32,32,3))

    # Input Layer
    model = Sequential()
    model.add(Conv2D(filters= 3, kernel_size=(3,3), padding='same', input_shape = (32,32,3)))

    # add tranfer learning layer
    model.add(vgg16_model)

    # Average Pooling
    model.add(GlobalAveragePooling2D())
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    # Full Connected Layer
    model.add(Dense(units = 512, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(Dense(units = 1, activation='sigmoid'))

    # loss = binary_crossentropy
    model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    return model
```

```python
model = build_model()
```
